```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([20,25,30,35,40,45,50,55,60,65])
y = np.array([0.5,1.8,4.8,5.5,6,6.5,7.8,9.4,9.8,10.9])

plt.plot(x,y,'*b')
```

```python
def SLR(x,y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    num = 0
    den = 0
    for i in range(len(x)):
        num += (y[i] - y_mean) * (x[i] - x_mean)
        den += (x[i] - x_mean) ** 2
    eps = 0.00000001
    a = num / (den + eps)
    b = y_mean - a * x_mean
    return a,b
```

```python
print(SLR(x,y))
slope, bias = SLR(x,y)
```

```python
x_new = np.linspace(0,100,100)
# y_pred = slope * x + bias
y_pred = slope * x_new + bias
plt.plot(x,y,'*b')
plt.plot(x_new,y_pred,'-r')
```

```python
def R2(a, b ,x,y):
    y_pred = a * x + b
    y_mean = np.mean(y)
    RSS = 0
    TSS = 0
    eps = 0.00000001
    for i in range(len(x)):
        RSS += (y[i] - y_pred[i]) ** 2
        TSS += (y[i] - y_mean) ** 2
    return 1 - (RSS / (TSS + eps))
```

```python
print(R2(slope,bias,x,y))
```

```python
def MLR(X,Y):
    Xt = X.T
    XtX = np.dot(Xt,X)
    # XtX = np.matmul(Xt,X)
    XtX_inv = np.linalg.inv(XtX)
    XtX_inv_Xt = np.dot(XtX_inv,Xt)
    # XtX_inv_Xt = np.matmul(XtX_inv,Xt)
    BETAs = np.dot(XtX_inv_Xt,Y)
    # BETAs = np.matmul(XtX_inv_Xt,Y)
    return BETAs
```

```python
X = np.array([x, x**2, x**3, np.ones(len(x))])
print(X.T)
```

```python
X = np.array([x, x**2, x**3, np.ones(len(x))])
y = np.array([0.5,1.8,4.8,5.5,6,6.5,7.8,9.4,9.8,10.9])
BETAs = MLR(X.T,y)
xnew = np.linspace(0,100,100)
Xnew = np.array([x_new, x_new**2, x_new**3, np.ones(len(xnew))])
# y_pred = np.dot(BETAs,X)
y_pred = np.dot(BETAs,Xnew)
# print(y_pred)
plt.plot(x,y,'*b')
plt.plot(xnew,y_pred,'-r')
```
