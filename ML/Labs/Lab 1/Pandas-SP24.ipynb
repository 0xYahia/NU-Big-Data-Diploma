* **rename target 1 to 'yes', and 0 to 'no'**
* **rename gender 1 to 'male', and 0 to 'female'**


````python
#Solution 1
table['gender'] = table['sex'].map({1: 'Male', 0: 'Female'})
table['attack'] = table['target'].map({1:'Yes' ,0: 'No'})
gender_attack=pd.crosstab(table.gender, table.attack, margins = True)
gender_attack
````

````python
#Solution 2
table['gender'] = table['sex'].map(lambda x: 'Male' if x==1 else 'Female')
table['attack'] = table['target'].map(lambda x: 'Yes' if x ==1 else 'No')
gender_attack = pd.crosstab(table.gender, table.attack, margins = True)
gender_attack
````

````python
gender_attack.loc['Female','Yes']/gender_attack.loc['All','Yes']
````

````python
gender_attack.iloc[0,1]/gender_attack.iloc[2,1]
````

***Exercise***

fbs: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)

* create new column 'bloodSugar' with values 'Yes' or 'No'

cp: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)

* convert chest pain column values to their corresponding meanings




````python
## code here
table['bloodSugar'] = table['fbs'].map(lambda x: 'Male' if x==1 else 'Female')
````

````python
table['cp'].value_counts()
````

````python
dict={0:'typical angina', 1:'atypical angina', 2:'non-anginal pain', 3:'asymptomatic'}
table['chestPain'] = table['cp'].map(lambda x: dict.get(x))
````

````python
# another solution
table['chestPain'] = table['cp'].map({0:'typical angina', 1:'atypical angina', 2:'non-anginal pain', 3:'asymptomatic'})
````
